```python
import matplotlib.pyplot as plt
import pandas as pd

df_org = pd.read_csv("homepap_cleaned.csv")
print(df_org.columns)
df = df_org

# Two sets of data
x = df['fosq_global']
y = df['cal_total']


# Plot the data as a line graph
plt.scatter(x, y, color='orange')

# Add labels to the axes and a title
plt.xlabel('fosq_global')
plt.ylabel('cal_total')
plt.title('Graph of fosq_global vs cal_total')

# Display the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('homepap_cleaned.csv')

# Create a histogram of age vs fosq_global
plt.hist2d(df['gender'], df['fosq_global'], bins=20, cmap='Blues')
plt.xlabel('Age')
plt.ylabel('FOSQ-Global')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file into a pandas dataframe
df = pd.read_csv('homepap_cleaned.csv')

# Create a barplot
plt.bar(df['dxhtn'], df['fosq_global'])

# Set the title and labels for the x and y axis
plt.title('Gender vs. FOSQ Global')
plt.xlabel('Gender')
plt.ylabel('FOSQ Global')

# Show the plot
plt.show()
```
